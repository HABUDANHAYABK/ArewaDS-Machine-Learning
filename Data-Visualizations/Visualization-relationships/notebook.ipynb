# Visualizing Relationships
Visualizing relations between variables helps uncover trends, patterns and potential correlations.

Visualizing relationships can be done using various types of plots, including:
  * Scatter plots
  * Line plots
  * Bubble plots
  * Pair plots (via Seaborn)

### `Scatter plots`
A scatter plots shows how two numerical variables relate. Each point represents an observation, with the x and y coordinates showing its values.
 * Ideal for correlation analysis
 * Not suitable for categorical comparisions

### `Use Cases`
 * Height vs Weigh Analysis
 * Age vs Income
 * Advertising budget vs Sales

### Example 1: Hours Studied vs Exam Score 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

hours = np.array([1, 2, 3, 4, 5, 6, 7, 8])
scores = np.array([50, 55, 60, 65, 70, 75, 80, 85])

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(hours, scores, color='teal')
ax.set_title("Hours Studied vs Exam Scores")
ax.set_xlabel("Hours Studies")
ax.set_ylabel("Exam Scores (%)")

plt.grid(True)

plt.show()
```

### Example2: Engine size vs Fuel Efficiency

```python
engine_size = [1.0, 1.5, 1.6, 2.0, 2.5, 3.0, 3.5]
mpg = [40, 38, 35, 30, 28, 22, 20]

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(engine_size, mpg, color='orange')
ax.set_title("Engine Size vs  MPG")
ax.set_xlabel("Engine Size (L)")
ax.set_ylabel("Miles per Gallon")
plt.grid(True)

plt.show()
```
