# Visualizing Relationships
Visualizing relations between variables helps uncover trends, patterns and potential correlations.

Visualizing relationships can be done using various types of plots, including:
  * Scatter plots
  * Line plots
  * Bubble plots
  * Pair plots (via Seaborn)

### `Scatter plots`
A scatter plots shows how two numerical variables relate. Each point represents an observation, with the x and y coordinates showing its values.
 * Ideal for correlation analysis
 * Not suitable for categorical comparisions

### `Use Cases`
 * Height vs Weigh Analysis
 * Age vs Income
 * Advertising budget vs Sales

### Example 1: Hours Studied vs Exam Score 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

hours = np.array([1, 2, 3, 4, 5, 6, 7, 8])
scores = np.array([50, 55, 60, 65, 70, 75, 80, 85])

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(hours, scores, color='teal')
ax.set_title("Hours Studied vs Exam Scores")
ax.set_xlabel("Hours Studies")
ax.set_ylabel("Exam Scores (%)")

plt.grid(True)

plt.show()
```

### Example2: Engine size vs Fuel Efficiency

```python
engine_size = [1.0, 1.5, 1.6, 2.0, 2.5, 3.0, 3.5]
mpg = [40, 38, 35, 30, 28, 22, 20]

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(engine_size, mpg, color='orange')
ax.set_title("Engine Size vs  MPG")
ax.set_xlabel("Engine Size (L)")
ax.set_ylabel("Miles per Gallon")
plt.grid(True)

plt.show()
```

### `Line Charts`
A line chart connects data points in a sequence - often used for time series or continuous data.
 * Best for trends over time
 * Doesn't show distributions or clusters
  
### `Use Cases`
 * Daily Temperature
 * Stock price over time
 * Sales over months

### Example 2: Monthly revenue trend

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
revenue = [1000, 1200, 1400, 1600, 1550, 1650]

fig, ax = plt.subplots(figsize=(7,7))
ax.plot(months, revenue, marker='o', color='navy')
ax.set_title("Monthly Revenue")
ax.set_xlabel("Months")
ax.set_ylabel("Revenue ($)")

plt.grid(True)

plt.show()
```

### Example 2: Website visit over 7 days

```python
days = ['Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sat', 'Sun']
visits = [120, 135, 150, 160, 155, 200, 180]

fig, ax = plt. subplots(figsize=(7,7))
ax.plot(days, visits, marker='*', linestyle='--', color= 'green')
ax.set_title("Website Traffic Over a week")
ax.set_xlabel("Days")
ax.set_ylabel("Number of Visits")

plt.grid(True)
plt.show()
```

### ` Bubble Chart`
> A bubble chart is a scatter plot where each point also has a size(3rd dimension), showing a third variable visually.
 * Great for showing impact with size
 * Can get cluttered with many points

### `Uses Cases`
 * Product sales vs profits vs volume
 * Population vs area vs gdp
 * Social Media engagements

### Example 3: Marketing budget vs sales vs conversion

```python
budget = [100, 200, 300, 400, 500]
sales = [10, 25, 30, 40, 55]
conversion = [20, 30, 40, 60, 80]

fig, ax =plt.subplots(figsize=(5, 5))
ax.scatter(budget, sales, s=conversion, alpha=0.5, c='purple')
ax.set_title("Marketing Spend vs Sales (Bubble = Conversion)")
ax.set_xlabel("Marketing Budget ($k)")
ax.set_ylabel("Sales ($k)")

plt.grid(True)

plt.show()
```

### Example 2: GDP VS Population vs Area 

```python
countries = ['A', 'B', 'C', 'D']
gdp = [1.2, 2.0, 2.5, 3.5]
area = [100, 150, 300, 200]
population = [30, 40, 50, 60]

fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(gdp, population, s=area, alpha=0.6, color='skyblue')
ax.set_title("GDP vs Population (Bubble = Area)")
ax.set_xlabel("GDP (Trillions)")
ax.set_ylabel("Population (Millions)")

plt.grid(True)
plt.show()
```
