# VISUALIZATIONS OF DISTRIBUTIONS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# set seed to ensure reproducibility
np.random.seed(43)

scores_tough_subject = np.random.beta(a=1, b=10, size=200)*100
#print(scores_tough_subject)
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(scores_tough_subject, bins = 20, color = 'blue', edgecolor = 'black')
ax.set_xlabel("Score")
ax.set_ylabel("Number of Students")
ax.set_title("Student Scores")

plt.grid(True)

plt.show()
```

## `Easy Exam`

* Left skewed distributions
* Scenario: An easy subject or generous grading where most students score high

```python
scores_easy_subject = np.random.beta(a= 5, b=1.5, size=200)*100
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(scores_easy_subject, bins = 20, color = 'blue', edgecolor='black')
ax.set_xlabel("Scores")
ax.set_ylabel("Number of Students")
ax.set_title("Students Scores on Easy Subject (Righ-Skewed)")

plt.grid(True)

plt.show()
```

## `Bimodal  Distributions`

* Two peaks in the `data`
* A subject with two student groups: `prepared` vs `unprepared`

```python
scores_mixed_ability = np.concatenate([np.random.normal(loc=50, scale=8, size=100),
                                      np.random.normal(loc=85, scale=5, size=100)])
scores_mixed_ability = np.clip(scores_mixed_ability, 0, 100)

fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(scores_mixed_ability, bins = 15, color='blue', edgecolor='black')
ax.set_title("Students scores (Bimodal): Mixed Ability Class")
ax.set_xlabel("Scores")
ax.set_ylabel("Number of Students")
plt.grid(True)

plt.show()
```

```python
# Simulate scores
scores_tough_subject = np.random.beta(a=1, b=4, size=200)*100
scores_normal_subject = np.random.normal(loc=75, scale=10, size=200)
scores_normal_subject = np.clip(scores_normal_subject, 0, 100)

#plot histogram
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(scores_tough_subject, bins = 15, alpha =0.7, label= 'Tough Subject (Beta)', edgecolor='black')
ax.hist(scores_normal_subject, bins =15, alpha=0.7, label ='Normal Subject (Normal)', edgecolor = 'black')
ax.set_title("Distributions of Students Scores")
ax.set_xlabel("Scores")
ax.set_ylabel("Number of Students")

plt.legend()
plt.grid(True)

plt.show()
```

## `BOX PLOTS`

* Summarize the class performance
* Find the median score
* Detect any very high or low score (outliers)
* Quick Compare this class to other later

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(1)

# Simulate students score around 70 -85
scores = np.random.normal(loc=75, scale= 10, size=100)
scores[:5] = np.random.uniform(30, 45, size=5)
scores = np.clip(scores, 0, 100)

print(scores)
```

```python
fig, ax = plt.subplots(figsize=(8, 5))
ax.boxplot(scores, vert= False)
ax.set_title("Box Plot: Final Exam Score (100 Students)")
ax.set_xlabel("Scores")
plt.grid(True)

plt.show()
```
