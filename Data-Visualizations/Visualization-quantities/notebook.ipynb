## `Data Visualization`

> The most powerfull tools for Data Scientist is the ability to visualize data and craft meaning insight from it. A well crafted graph or plot can ease to draw a conclusion from the `Data`

### `Bar Chart`
A bar-chart is a plot use to visualize categorical data. With `matplotlib` library all the visualizations are made easier. 

```python
import matplotlib.pyplot as plt

nations = ["Nigeria", "Niger", "Chadi", "Ghana"]
pop = [200, 50, 31, 61]

#customize color
color_label = {
    "Nigeria": "green",
    "Niger": "blue",
    "Chadi": "red",
    "Ghana": "black"
}
colors = [color_label[country] for country in nations]
plt.figure(figsize=(8, 4))

plt.bar(nations, pop, color= colors)
plt.title("Some African Countries with their Populations")
plt.xlabel("Countries")
plt.ylabel("Population in millions")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

birds = pd.read_csv(r"c:/users/danhaya/arewads/arewads-fellowship/arewads-machine-learning/data/birds.csv")
birds.head()
```

```python
birds.plot(kind='scatter', x= 'MaxLength', y= 'Order', figsize=((12, 8)))
plt.title("Max Length per Order")
plt.ylabel("Order")
plt.xlabel("Max Length")

plt.show()
```

```python
birds['MaxBodyMass'].plot(kind='hist', figsize=((12, 8)))
plt.show()
```

```python
birds.plot(kind = 'scatter', x = "Order", y= "MaxLength", figsize=((12, 8)))
plt.title("Maximum Length per Order")
plt.ylabel("Max Length")
plt.xlabel("Order")
plt.xticks(rotation = 90)

plt.show()
```

```python
filteredbirds = birds[(birds['MaxBodyMass'] > 1) & (birds["MaxBodyMass"] < 500)]
print(len(birds), len(filteredbirds))

filteredbirds['MaxBodyMass'].plot(kind ='hist', bins = 40, figsize=((12, 8)))
plt.title("Distribution of Body Mass")

plt.show()
```
