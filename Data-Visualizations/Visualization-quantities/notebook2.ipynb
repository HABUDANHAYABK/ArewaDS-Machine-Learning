```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sales_dict = {
    "products" : ["Laptpos", "Phones", "Tablets"],
    "Sales": [1000, 5000, 10000]
}

df = pd.DataFrame(sales_dict)

fig, ax = plt.subplots()
ax.bar(df['products'], df['Sales'], color = ['b', 'g', 'r'])
ax.set_xlabel("Products")
ax.set_ylabel("Sales ($)")
ax.set_title("Sales by Products")

plt.show()
```

```python
data = {
    "model": ["Sedan", "SUV", "Truck", "Hatchback"],
    "Units sold": [150, 220, 180, 90]
}
df = pd.DataFrame(data)

fig, ax = plt.subplots()
plt.Figure(figsize=((5, 5)))
ax.bar(df['model'], df['Units sold'], color = ['blue', 'green', 'black', 'orange'])
ax.set_xlabel("Models")
ax.set_ylabel("Units Sold")
ax.set_title("Model by Units sold")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Savings": [100, 150, 180, 250, 230, 300]
}

savings_df = pd.DataFrame(data)

fig, ax = plt.subplots()
ax.plot(savings_df["Month"], savings_df['Savings'], marker = 'o', linestyle= '-', color = 'purple')
ax.set_xlabel("Months")
ax.set_ylabel("Savings")
ax.set_title("Savings by Month")
plt.grid(True, alpha = 0.7)

plt.show()
```
