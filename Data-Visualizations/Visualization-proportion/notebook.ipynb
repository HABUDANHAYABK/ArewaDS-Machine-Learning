# VISUALIZATION S OF PROPORTIONS

Visualizing proportions helps us understand how parts contribute to a whole. We'll explore:
* Pie Charts
* Donut Charts
* Stacked Bar Charts
* Percentage Bar Charts

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

brands = ["Apple", "Samsung", "Xiaomi", "Others"]
shares = [40, 30, 20, 10]

fig, ax = plt.subplots(figsize=(7, 8))
ax.pie(shares, labels=brands, autopct='%1.1f%%',startangle=140)
ax.set_title("Smartphone Market Share")

plt.show()
```

### `Example 2:`
Survey responses to "Prefered transport"


```python
modes = ["Car", "Bike", "Bus", "Walk"]
counts = [50, 15, 20, 15]

fig, ax = plt.subplots(figsize= (7, 7))
ax.pie(counts, labels=modes, autopct= lambda p: f'{p:.0f}%\n({p*sum(counts)/100:.0f})')
ax.set_title("Prepared Transport Mode")

plt.show()
```

## ` Donut Chart`
 > A donut chart is a pie chart with a hole in the middle. it's usually clearner  and allows room for central labels or values.

 * Good for dashboards and reports
 * Offers a sleek alternative to pie charts.
  
  ### `Use Cases`:
  * Time allocation
  * Revenue/ Source analysis
  * Distribution summaries in presentations

```python
categories = ["Product A", "Product B", "Services", "Others"]
values = [120, 80, 100, 50]

fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(values, labels= categories, autopct="%1.1f%%",
                                  startangle=90, wedgeprops={'width': 0.4})
ax.set_title("Startup Revenue Breakdown")
plt.show()
```

Example 2: Time Allocation in a workday

```python
activities = ["work", "Meetings", "Breaks", "Learning"]
hours = [6, 2, 1, 3]

fig, ax = plt.subplots(figsize=(6, 6))
ax.pie(hours, labels=activities, autopct='%1.1f%%', startangle=90, wedgeprops={'width': 0.5})
ax.set_title("Daily Time Allocation")

plt.show()
```
