# VISUALIZATION S OF PROPORTIONS

Visualizing proportions helps us understand how parts contribute to a whole. We'll explore:
* Pie Charts
* Donut Charts
* Stacked Bar Charts
* Percentage Bar Charts

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

brands = ["Apple", "Samsung", "Xiaomi", "Others"]
shares = [40, 30, 20, 10]

fig, ax = plt.subplots(figsize=(7, 8))
ax.pie(shares, labels=brands, autopct='%1.1f%%',startangle=140)
ax.set_title("Smartphone Market Share")

plt.show()
```

### `Example 2:`
Survey responses to "Prefered transport"


```python
modes = ["Car", "Bike", "Bus", "Walk"]
counts = [50, 15, 20, 15]

fig, ax = plt.subplots(figsize= (7, 7))
ax.pie(counts, labels=modes, autopct= lambda p: f'{p:.0f}%\n({p*sum(counts)/100:.0f})')
ax.set_title("Prepared Transport Mode")

plt.show()
```
